```python
contagem = df.groupby(['ProvCand', 'Sexo']).size
```

```python
print(contagem)
```

```python
provincias = contagem['ProvCand'].unique()

for ProvCand in provincias:
    dados_provincia = contagem[contagem['ProvCand'] == ProvCand]
    generos = dados_provincia['Sexo']
    contagens = dados_provincia['Contagem']


    plt.figure()
    plt.pie(contagens, labels=generos, autopct='%1.0f%%')
    plt.axis('equal')
    plt.title('Distribuição de Gênero - Província: ' + ProvCand)

plt.show()
```

```python
grupo = df.groupby(['ProvCand', 'Sexo']).size().unstack()
```

```python
grupo.plot(kind='bar', stacked=True)
plt.xlabel('Províncias')
plt.ylabel('Gênero')
plt.title('Distribuição de Gênero por Província')
plt.legend(title='Sexo')
plt.show()
```
